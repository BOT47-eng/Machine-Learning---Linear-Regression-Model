```python
import numpy  as np
from pandas import DataFrame as df
from os import system
```

Single Feature Training 

```python
SIZE = 5

x0 = np.ones(SIZE)
x1 = np.arange(SIZE)
y = np.arange(SIZE) +   2
data = df({'X0' : x0 , 'X1' : x1  ,'Y' : y}) # Where X1 is my feature 1 

# For each feature i need weights 
wieghtLen = len(data.columns) - 1 # - 1 is for the Y
wieghts = {f'X{st}Theta' : 0 for st in range(wieghtLen)} # Starting weight is 0 for each feature and + 1 for the  x0Theta feature biased value
learningRate = 0.01  

"""
M = Number of training Examples 
Original Mean Squared Error Loss Function = (From i - M Sum ((hTheta(x(i)) - y(i))^2)) / 2 * M  # we also divide by 1/2 just sort of convention so when we take derievative it automatically gets removed.

hTheta(x(i)) = for Each weight starting from 1 Multiply it by the respective feature each point of the dataset  and add all the weights.

After Feature Derievative = (From  i to M Sum (hTheta(x(i)) - y(i)) * x(i)) / M

After the x0Theta where x0 === 1 as a convention = (From  i to M Sum (hTheat(x(i)) - y(i))) / M


Also Think of x0Theta as  the b 


CASE FOR EACH FEATURE STARTING FROM x1Theta :

First for each weight multiply it with all the data points of the feature, in this case we have also have to take the square of x(i)^2 when multiplying and add. 
Then Multiply that specific feature data points with the x0Theta/b and add
Divide it with the number of data point to get the specific weight

CASE FOR THE FEATURE of x0Theta : 

We Multiply simply with the and add without any square thing.

"""


def RateofChange(wieghts , data , currentWeightKey , totalDataPoints) :
    # dJ_dkey = SUM(((EACH WEIGHT * BY ITS DATA SET AND ADD ALL WEIGHTS THE -  Y)  *  BY ITS DATA SET )  / TOTAL SIZE )
    dJ = 0
    for keys in wieghts :
        dJ += sum(wieghts[keys] * data[str(keys[:2:1])] * data[currentWeightKey[:2:1]])
    dJ -= sum(data['Y'] * data[currentWeightKey[:2:1]])
    dJ /= totalDataPoints 
    return dJ

while(True):
    for keys in wieghts : 
        wieghts[keys] = wieghts[keys] - learningRate *  RateofChange(wieghts , data, keys , SIZE) 
        system("clear")
        print(wieghts)
```
